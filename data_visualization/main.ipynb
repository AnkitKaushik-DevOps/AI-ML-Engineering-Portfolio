```python
import matplotlib.pyplot as plt
```

```python
# data
x = [1, 2, 3, 4, 5]
y = [10, 25, 35, 45, 65]
```

```python
plt.plot(x, y)
plt.show()
```

```python
plt.plot(x, y, 'ro-')
plt.show()
```

```python
plt.scatter(x, y)
plt.plot(x, y)
plt.show()
```

```python
# Import Numpy

import numpy as np
```

```python
# Find the Correlation
'''
Interpretation:
r ≈ 1 → strong linear relationship
r ≈ 0 → no linear relationship

If r > 0.9, linear regression is usually reasonable.

'''
r = np.corrcoef(x, y)[0, 1]
print(r)
```

```python
'''
Finding Residuals(esiduals are the differences between the observed values and the values predicted by a model.)

Residual=y^observed​−y^predicted​

Key Rule:
Residuals should be randomly scattered
If you see a pattern (curve, trend) → linear model is inappropriate

'''

coeffs = np.polyfit(x, y, 1)
y_pred = np.polyval(coeffs, x)

residuals = y - y_pred
```

```python
plt.scatter(x, residuals)
plt.axhline(0)
```

```python
'''
R² Score (Goodness of Fit)

Interpretation:
R² ≈ 1 → good fit
R² < 0.8 → weak linear fit

'''

ss_res = np.sum((y - y_pred)**2)
ss_tot = np.sum((y - np.mean(y))**2)

r2 = 1 - (ss_res / ss_tot)
print(r2)
```
